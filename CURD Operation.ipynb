```python
def updated_record(df, invoice_id, column_name, new_value):
    
    """This function update records based on the Invoice ID.
        # Loc is used to Locate the row(s) in the DataFrame 
             where 'Invoice_ID' matches the given invoice_id."""
    
    df.loc[df['Invoice_ID'] == invoice_id, column_name] = new_value
    return df    # it return updated df2
```

### DELETE: 

Remove specific records from the dataset.

```python
def delete_record(df, Invoice_ID):     # It remove only 1 record by using unique identifier i.e Invoice_ID
    global df1
    df1 = df[df['Invoice_ID'] != Invoice_ID]        #it removes the records of given Invoice id
    return df1
    
```

```python
def delete_record(df,column, value):        # It remove multiple records on the basis of value which we are given
    global df2                                # golbal the dataframe for using dataframe out of the code
    df2 = df[df[column] != value]          # it create new dataframe contain records other than specific value which 
                                                             #we given in def function
    return df2
```
